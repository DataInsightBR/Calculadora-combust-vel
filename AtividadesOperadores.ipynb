```python
# Programa: Soma de dois números
# Descrição: Lê dois números digitados pelo usuário e mostra a soma.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados (input sempre retorna texto)
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

# Processamento (cálculo da soma)
soma = numero1 + numero2

# Saída de dados (exibição do resultado)
print("A soma é:", soma)
```

```python
# Programa: Identificador de tipo de variável
# Descrição: Lê um valor digitado pelo usuário e exibe o tipo do dado.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados (sempre retorna string)
valor = input("Digite algo: ")

# Saída mostrando o tipo da variável
print("O tipo da variável é:", type(valor))
"""
A função input() sempre retorna texto (str) porque tudo que o usuário
digita no teclado chega ao programa como uma sequência de caracteres.

O Python não “adivinha” o tipo do dado
Ele apenas captura exatamente o que foi digitado

Porque isso evita erros e ambiguidades.
Se o Python tentasse adivinhar automaticamente, poderia interpretar errado.

Você precisa converter explicitamente o tipo:
"""

numero_inteiro = int(input("Digite um número inteiro: "))
numero_decimal = float(input("Digite um número decimal: "))

print(type(numero_inteiro))  # <class 'int'>
print(type(numero_decimal))  # <class 'float'>
```

```python
# Programa: Antecessor e Sucessor
# Descrição: Lê um número e mostra seu antecessor e sucessor.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
numero = int(input("Digite um número inteiro: "))

# Cálculo
antecessor = numero - 1
sucessor = numero + 1

# Saída
print("Número digitado:", numero)
print("Antecessor:", antecessor)
print("Sucessor:", sucessor)
```

```python
# Programa: Dobro, Triplo e Quadrado
# Descrição: Lê um número digitado pelo usuário e mostra o dobro, triplo e o valor ao quadrado.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
numero = float(input("Digite um número: "))

# Cálculos
dobro = numero * 2
triplo = numero * 3
quadrado = numero ** 2

# Saída
print("Número digitado:", numero)
print("Dobro:", dobro)
print("Triplo:", triplo)
print("Ao quadrado:", quadrado)
```

```python
# Programa: Cálculo de Média
# Descrição: Lê duas notas de um aluno, calcula a média e exibe o resultado.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))

# Cálculo da média
media = (nota1 + nota2) / 2

# Saída
print("Notas:", nota1, "e", nota2)
print("Média do aluno:", media)
```

```python
# Programa: Conversor de Medidas
# Descrição: Lê um valor em metros e mostra o equivalente em centímetros e milímetros.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
metros = float(input("Digite o valor em metros: "))

# Conversões
centimetros = metros * 100
milimetros = metros * 1000

# Saída
print("Valor em metros:", metros)
print("Em centímetros:", centimetros)
print("Em milímetros:", milimetros)
```

```python
# Programa: Conversor de Real para Dólar
# Descrição: Lê um valor em reais e mostra o equivalente em dólares.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Cotação usada: 1 USD = R$ 5.02 (valor aproximado)
# Data: 2026

# Entrada de dados
reais = float(input("Digite o valor em reais (R$): "))

# Cotação do dólar (valor fixo para o programa)
cotacao_dolar = 5.02

# Conversão
dolares = reais / cotacao_dolar

# Saída
print("Valor em reais: R$", reais)
print("Valor em dólares: US$", dolares)
```

```python
# Programa: Cálculo de Tinta para Parede
# Descrição: Lê a altura e largura de uma parede e calcula a quantidade de tinta necessária.
# Consideração: 1 litro de tinta pinta 10 m² por demão.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
altura = float(input("Digite a altura da parede (em metros): "))
largura = float(input("Digite a largura da parede (em metros): "))

# Cálculo da área
area = altura * largura

# Cálculo da quantidade de tinta (1L cobre 10m²)
tinta = area / 10

# Saída
print("Área da parede:", area, "m²")
print("Quantidade de tinta necessária:", tinta, "litros")
```

```python
# Programa: Cálculo de Desconto
# Descrição: Lê o preço de um produto e mostra o valor com 5% de desconto.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
preco = float(input("Digite o preço do produto (R$): "))

# Cálculo do desconto (5%)
desconto = preco * 0.05
preco_final = preco - desconto

# Saída
print("Preço original: R$", preco)
print("Desconto (5%): R$", desconto)
print("Preço com desconto: R$", preco_final)
```

```python
# Programa: Cálculo de Aumento Salarial
# Descrição: Lê o salário de um funcionário e mostra o valor com 15% de aumento.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
salario = float(input("Digite o salário (R$): "))

# Cálculo do aumento (15%)
aumento = salario * 0.15
novo_salario = salario + aumento

# Saída
print("Salário atual: R$", salario)
print("Aumento (15%): R$", aumento)
print("Novo salário: R$", novo_salario)
```

```python
# Programa: Área do Círculo
# Descrição: Lê o raio de um círculo e calcula sua área.
# Fórmula: área = π * raio²
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

import math  # Biblioteca para usar o valor de pi (π)

# Entrada de dados
raio = float(input("Digite o raio do círculo: "))

# Cálculo da área
area = math.pi * (raio ** 2)

# Saída
print("Raio:", raio)
print("Área do círculo:", area)
```

```python
# Programa: Média de Preços de Combustível
# Descrição: Lê três preços e calcula a média aritmética.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Entrada de dados
preco1 = float(input("Digite o primeiro preço (R$): "))
preco2 = float(input("Digite o segundo preço (R$): "))
preco3 = float(input("Digite o terceiro preço (R$): "))

# Cálculo da média
media = (preco1 + preco2 + preco3) / 3

# Saída
print("Preços informados:", preco1, preco2, preco3)
print("Média dos preços: R$", media)
```

```python
# Programa: Lista de Compras
# Descrição: Lê nome, quantidade e preço de 3 produtos e calcula o total da compra.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Produto 1
nome1 = input("Digite o nome do produto 1: ")
qtd1 = float(input("Digite a quantidade do produto 1: "))
preco1 = float(input("Digite o preço do produto 1 (R$): "))

# Produto 2
nome2 = input("Digite o nome do produto 2: ")
qtd2 = float(input("Digite a quantidade do produto 2: "))
preco2 = float(input("Digite o preço do produto 2 (R$): "))

# Produto 3
nome3 = input("Digite o nome do produto 3: ")
qtd3 = float(input("Digite a quantidade do produto 3: "))
preco3 = float(input("Digite o preço do produto 3 (R$): "))

# Cálculo dos totais por produto
total1 = qtd1 * preco1
total2 = qtd2 * preco2
total3 = qtd3 * preco3

# Total geral
total_geral = total1 + total2 + total3

# Saída
print("\n=== Lista de Compras ===")
print(nome1, "- Quantidade:", qtd1, "- Total: R$", total1)
print(nome2, "- Quantidade:", qtd2, "- Total: R$", total2)
print(nome3, "- Quantidade:", qtd3, "- Total: R$", total3)

print("\nTotal da compra: R$", total_geral)
```

```python
# Programa: Comparações Lógicas
# Descrição: Avalia expressões com valores fixos e mostra verdadeiro ou falso.
# Fonte/Apoio: Código gerado com auxílio de IA (ChatGPT - OpenAI)
# Data: 2026

# Valores definidos
a = 4
b = 10
c = 5
d = 1
f = 5

# Comparações
print("a == c:", a == c)
print("a < b:", a < b)
print("d > b:", d > b)
print("c != f:", c != f)
print("a == b:", a == b)
print("c < d:", c < d)
print("b > a:", b > a)
print("c >= f:", c >= f)
print("f >= c:", f >= c)
print("c <= c:", c <= c)
print("c <= f:", c <= f)
```
